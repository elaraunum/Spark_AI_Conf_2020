```

# required libraries
import gym
import numpy as np
from gym import spaces
np.random.seed(1234)


class GridWorldEnvironment(gym.Env):
  '''
  This class describes the simple version of Deal or No Deal environment
  '''
  
  UP = 0
  DOWN = 1
  RIGHT = 2
  LEFT = 3
  
  def __init__(self):
    
    # define the immediate reward, number of actions and number of states
    
    self.is_done = False
    self.reward = -1
    
    # number of actions
    self.na = 4
    # number of states
    self.ns = 16
    #initial state
    self.state = 0
    
    self.action_space = spaces.Discrete(self.na)
    self.observation_space = spaces.Discrete(self.ns)


  # define reset() method. This method reset the environment.
  def reset(self):
    return self._reset()
  
  # define step() method. This method describes how the environment responds. 
  def step(self, action):
    return self._step(action)
  
  def set_state(self, state):
    self.state = state
  
  def _step(self, action):
  
    
    # going up
    if action == self.UP:
      
      
      
      if self.state in range(4) and self.state != 0:
        self.state = self.state
        self.reward = -1
        self.is_done = False

        
      elif self.state == 0:
        self.state = self.state
        self.reward = 0
        self.is_done = True
      
      else:
        self.state -= 4
        self.reward = -1
        self.is_done = False
        if self.state == 0 or self.state == 15:
          self.is_done = True
        
    # going down    
    if action == self.DOWN:
      
      
      
      if self.state in range(12, 16) and self.state != 15:
        self.state = self.state
        self.reward = -1
        self.is_done = False
        
      elif self.state == 15:
        self.state = self.state
        self.reward = 0
        self.is_done = True
         
      else:
        self.state += 4
        self.reward = -1
        self.is_done = False
        if self.state == 0 or self.state == 15:
          self.is_done = True
        
        
    # going right   
    if action == self.RIGHT:
      
      
      
      if self.state in range(3, 16, 4) and self.state != 15:
        self.state = self.state
        self.reward = -1
        self.is_done = False
        
      elif self.state == 15:
        self.state = self.state
        self.reward = 0
        self.is_done = True
  
      
      else:
        self.state += 1
        self.reward = -1
        self.is_done = False
        if self.state == 0 or self.state == 15:
          self.is_done = True
        
        
    # going left    
    if action == self.LEFT:
      
      
      if self.state in range(0, 13, 4) and self.state != 0:
        self.state = self.state
        self.reward = -1
        self.is_done = False
        
      elif self.state == 0:
        self.state = self.state
        self.reward = 0
        self.is_done = True
        
        
      else:
        self.state -= 1
        self.reward = -1
        self.is_done = False
        if self.state == 0 or self.state == 15:
          self.is_done = True
  
    
  
    
    return [self.state, self.reward, self.is_done, {}]
  
  # this method reset the episode 
  def _reset(self):
    
    return 0 
```
