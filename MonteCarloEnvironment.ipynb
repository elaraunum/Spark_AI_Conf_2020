```

import random
import numpy as np
from statistics import mean
def monte_carlo(number_episodes, start_state = 'c'):
  '''
  This function generates multiple episodes for the random walk problem. 
  input: starting state where the random walk starts
  output: list of episodes 
  '''
  np.random.seed(1234)
  transitions = []
  # set up all states
  states = ['t1','a', 'b', 'c', 'd', 'e', 't2']
  
  #create samples of episodes
  for i in range (number_episodes):
    local_transitions = []
    start_index = states.index(start_state)
    local_transitions.append(start_index)
    while True:
      pick_action = np.random.randint(2, size = 1)[0]
      # go left
      if pick_action == 0:
        index = start_index - 1
      else:
        index = start_index + 1
      start_index = index
      local_transitions.append(index)
      if states[index] == 't1' or states[index] == 't2':
        transitions.append(local_transitions)
        break
    
  return(transitions) 
```
