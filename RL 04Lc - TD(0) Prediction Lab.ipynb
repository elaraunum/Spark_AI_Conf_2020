# Model-free Prediction (TD(0)) Lab

## ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) In this lab you learn:<br>
 - TD(0)

### Problem statement ###

Consider the random walk problem on a straight line (1D) with 7 states. States t1, a, b, c, d, e, t2. Assume two ends states, t1 and t2, are terminal states. If you end up there, the episode is over. If you reach t2, you are rewarded $100. If you reach t1, you are rewarded $0. There is a 50% chance of going right or left when you are at state a, b, c, d and e.

### Exercise ###
Assume you start at state c and \\(\gamma = 1, \alpha = 0.1\\). Use TD(0) to estimate the value of each state.

```
%run "./helper/MonteCarloEnvironment"
```

```
# ANSWER
def td_0(episodes, alpha=0.1, gamma=1.0):
  """This function calculate TD(0)."""
  
  reward_t2 = 100
  reward_t1 = 0
  values = [0, 0.5, 0.5, 0.5, 0.5, 0.5, 0]
  rewards = [reward_t1, 0, 0, 0, 0, 0, reward_t2]
  
  
  # Updates the values of states using TD(0)
  
  for i in range(len(episodes)):
    for j in range(len(episodes[i])):
      if episodes[i][j] != 0 and episodes[i][j] != 6:
        values[episodes[i][j]] =  values[episodes[i][j]] + alpha * (rewards[episodes[i][j+1]] + gamma * values[episodes[i][j+1]] - values[episodes[i][j]])
      else:
        break
  return values
```

```
# Test your code
episodes = monte_carlo(50000)
value = td_0(episodes) 
value_expected = [0.0, 15.0, 28.0, 41.0, 62.0, 82.0, 0.0] 
np.testing.assert_array_almost_equal(value, value_expected, err_msg = "The values are incorrect", decimal = 0)
```
