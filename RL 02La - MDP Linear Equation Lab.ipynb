# Linear Equation For Value Function

## ![Spark Logo Tiny](https://files.training.databricks.com/images/105/logo_spark_tiny.png) In this lab you learn:<br>
 - How to calculate the value function for a 3-state problem

### 3-state problem ###
For the following graph:
* Assume the return of state **\\(S, R, F\\) are -1, -3, 1**, respectively. Write a program to calculate the value of each state.

![MDPs](https://files.training.databricks.com/images/rl/mdps.png)
<br>

**Hint:** Use this formula
$$\begin{aligned} \\\ v^{k+1} &= \sum\_{a\in A} \pi(a\bigm\vert s) (R^a\_{s}+\gamma \sum\_{s'\in S} \rho\_{ss'}^a v\_{\pi}(s') \\\
&= R^{\pi} + \gamma \rho^{\pi}v^{k} \end{aligned}$$

Mote: You will need to use numpy for most of the computations. See [here](https://docs.scipy.org/doc/numpy/reference/index.html#reference) for numpy doc.

```
# ANSWER
import numpy as np

# Define a probability transition matrix
gamma_low = 0.1
gamma_medium = 0.5
gamma_high = 0.9

# Define transition probability matrix and reward vector
transition_matrix = np.matrix("0.5, 0.3, 0.2; 0.75, 0.1, 0.15; 0.2, 0.1, 0.7")
reward = np.matrix("-1; -3; 1")

# Solve a linear equation to find values of each state.
v_gamma_low = np.linalg.inv(np.identity(3) - gamma_low*transition_matrix)*reward
v_gamma_medium = np.linalg.inv(np.identity(3) - gamma_medium*transition_matrix)*reward
v_gamma_high = np.linalg.inv(np.identity(3) - gamma_high*transition_matrix)*reward

#Print the results
print(f"The value of states for gamma = 0.1 is \n {v_gamma_low}")
print(f"The value of states for gamma = 0.5 is \n {v_gamma_medium}")
print(f"The value of states for gamma = 0.9 is \n {v_gamma_high}")
```
